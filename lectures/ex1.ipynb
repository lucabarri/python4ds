```python
# generate 2 random arrays of dimention d, declare d as an int and set it equal to 5
# use np.random.rand() to generate 2 arrays of dimention d
# print the arrays

import numpy as np

d = 3
n = 5 # first set
m = 8 # second set

# generate a random matrix d x n

X = np.random.rand(n, d)
Y = np.random.rand(m, d)


distances = np.zeros((n, m))

for i in range(n):
    for j in range(m):
        for k in range(d):
            distances[i, j] += (X[i, k] - Y[j, k])**2

        distances[i, j] = (distances[i, j])**0.5
    print(distances[i])

```
